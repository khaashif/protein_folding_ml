```python
import taichi as ti
import math
from matplotlib import pyplot as plt
import numpy as np

# norm_range = 0.03
# step_size = 0.95
# too_close_pen = 5
# overlap_pen = 5
# beta = 0.1
# step_size_beta = 1.05


# too_close_penalty_list = list(range(1, 30, 5)) # 5
too_close_penalty_list=[0.001, 0.01, 0.1, 0.5, 1]
too_close_penalty_list=[0.1, 1]
# overlap_penalty_list = list(range(1, 30, 5)) # 5
overlap_penalty_list=[0.001, 0.01, 0.1, 0.5, 1] # 5
overlap_penalty_list = [0.1, 1]
step_size_close_list = [1.001, 1.01]
step_size_overlap_list = [1.001, 1.01]
# alpha_list = list(range(5, 50, 250)) # 50
alpha_list = [0.01, 0.05, 0.1, 1, 5, 10]# alpha 0.01 converges to local minima
alpha_list = [0.1, 5]
# step_size_list = list(np.arange(0.8, 1, 0.02)) # 0.05
step_size_list = [0.95, 0.9]
# beta_list = list(np.arange(0.01, 1.5, 0.05)) # 0.2
beta_list = [0.0001, 0.001, 0.01, 0.1]
beta_list = [0.001, 0.1]
# step_size_beta_list = list(np.arange(1.005, 1.1, 0.01)) # 0.05
step_size_beta_list = [1.0001, 1.001, 1.01, 1.05]
step_size_beta_list = [1.001]
# norm_range_list =  list(np.arange(0.04, 0.1, 0.3)) # 0.05
norm_range_list = [0.01, 0.05]

# gamma_list = [0.8, 0.9]

#Loop through all possible parameter combinations

#Note: Hyperparameter tuning for loops can be removed if undesired
for a in range(len(too_close_penalty_list)):
    for b in range(len(overlap_penalty_list)):
        for c in range(len(alpha_list)):
            for d in range(len(beta_list)):
                for e in range(len(step_size_list)):
                    for f in range(len(step_size_beta_list)):
                        for g in range(len(norm_range_list)):
                            for h in range(len(step_size_close_list)):
                                for i in range(len(step_size_overlap_list)):
                            
                                    # Convert parameters to be tuned into floats
                                    too_close_pen = float(too_close_penalty_list[a])
                                    overlap_pen = float(overlap_penalty_list[b])
                                    alpha = float(alpha_list[c])
                                    beta = float(beta_list[d])
                                    step_size = float(step_size_list[e])
                                    step_size_beta = float(step_size_beta_list[f])
                                    norm_range = float(norm_range_list[g])
                                    step_size_close = float(step_size_close_list[h])
                                    step_size_overlap = float(step_size_overlap_list[i])
                                     
                                    ti.init()
            
                                    #Instantiate variables
                                    U = ti.field(float, (), needs_grad=True)
                                    aa="HHPPHHPHHPPHHPHHPPHH"
                                    N = len(aa)
                                    seq = ti.Vector.field(1, float, N)
                                    theta = ti.Vector.field(1, float, N-1, needs_grad=True)  # angle array (transformed NEWS string)
                                    vel = ti.Vector.field(1, float, N-1)
                                    current_position = ti.Vector.field(2, float, 1)
                                    coords = ti.Vector.field(2, float, N, needs_grad=True)
                                    theta_conversion = ti.Vector.field(2, float, 360, needs_grad=True)

                                    # compute_U is the energy function 
                                    @ti.kernel
                                    def compute_U():
                                        #convert angular path into coordinates
                                        for i in range(1,N):
                                            coords[i] = coords[i-1] + [ti.cos(theta[i-1])[0], ti.sin(theta[i-1][0])]
                                        
                                        # loop through all pairs of amino acids
                                        for i, j in ti.ndrange(N, N):
                                            
                                            r = coords[i] - coords[j]
                                            
                                            # All penalties or rewards to the energy score are multiplied by a factor of 0.5
                                            #  as all pairs are repeated in for loop, e.g. i,j is the same pair as j,i
                                            
                                            #if statements to determine nature of bond between amino acid i and j
                                            if 1-norm_range <= r.norm(0) and r.norm(0) <= 1+norm_range: # if displacement = 1
                                                if abs(i-j)>1.0: # if not linked by chain
                                                    if seq[i][0] + seq[j][0] == 2: # if HH bond
                                                        U[None] -= 0.5 * abs(r.norm(1e-3)) * 3 # 
                                            elif r.norm(0) > norm_range and r.norm(0) < 1-norm_range: # if displacement between 0 and 1
                                                if i!=j: # if not itself
                                                    U[None] += 0.5 * 1 * (1/r.norm(1e-3))
                                            elif -norm_range <= r.norm(0) and r.norm(0) <= norm_range: # if displacement = 0 (overlap)
                                                if i!=j: # if not itself
                                                    U[None] += 0.5 * overlap_pen


                                        #penalty for non canonical theta values (0, 90, 180, 270)
                                        for i in theta:
                                            U[None] += beta * min((theta[i][0]-0)**2, (theta[i][0]-float(0.5*math.pi))**2, (theta[i][0]-float(math.pi))**2, (theta[i][0]-float(1.5*math.pi))**2)

                                    #gradient descent step, calculates next theta value 
                                    @ti.kernel
                                    def advance():
                                        for i in theta:
                                            theta[i] = theta[i] - (alpha*theta.grad[i])
                                            theta[i] = theta[i] % float(2*math.pi)
                                            
                                                    # commented out momentum implementation
        #                                         vel[i] = gamma*vel[i][0] + alpha*theta.grad[i]
        #                                         theta[i] = theta[i] - vel[i]

                                    #initialize sequence of amino acids and generate random angular path, theta
                                    @ti.kernel
                                    def init():
                                        coords[0][0], coords[0][1] = 0,0
                                        for i in theta:
                                            theta[i][0] = float(2*math.pi)*ti.random() # some value between 0 and 360
                                        # hardcoded numerical representation of sequence, H=1, P=0
                                        seq[0] = [1]
                                        seq[1] = [1]
                                        seq[2] = [0]
                                        seq[3] = [0]
                                        seq[4] = [1]
                                        seq[5] = [1]
                                        seq[6] = [0]
                                        seq[7] = [1]
                                        seq[8] = [1]
                                        seq[9] = [0]
                                        seq[10] = [0]
                                        seq[11] = [1]
                                        seq[12] = [1]
                                        seq[13] = [0]
                                        seq[14] = [1]
                                        seq[15] = [1]
                                        seq[16] = [0]
                                        seq[17] = [0]
                                        seq[18] = [1]
                                        seq[19] = [1]


                                    Us=[]
                                    coords_history=[]
                                    init()
        #                             gui = ti.GUI('Protein Folding',512)
                                    print("close_pen:", too_close_pen, ", overlap_pen:", overlap_pen, ", alpha:", alpha_list[c], 
                                      ", beta:", beta_list[d], ", alpha_step_size:", step_size, ", beta_step_size:", step_size_beta,
                                     ", norm_range:", norm_range, ", step_size_close:", step_size_close, ", step_size_overlap:", step_size_overlap)

                                    # 2000 iterations for differentiable algorithm
                                    # This for loop essentialy starts the optimization process, 
                                    #   recording structure and energy score at each iteration.
                                    
                                    for k in range(2000):
                                        with ti.Tape(U):
                                            compute_U()
                                        if k==0:
                                            min_coords = coords.to_numpy()
                                            # min_coords = coords.to_numpy()
                                            plt.scatter(min_coords.transpose()[0], min_coords.transpose()[1])
                                            plt.plot(min_coords.transpose()[0], min_coords.transpose()[1])
                                            plt.yticks(range(-8,8))
                                            plt.xticks(range(-8, 8))
                                            plt.xlim(-10, 10)
                                            plt.ylim(-10, 10)
                                            plt.gca().set_aspect('equal', adjustable='box')
                                            plt.title('Initial State')
                                            plt.show()
                                        Us.append(U[None])
                                        coords_history.append(coords)
                                        alpha = alpha * step_size
                                        beta = beta * step_size_beta
                                        too_close_pen = too_close_pen*step_size_close
                                        too_close_overlap = overlap_pen*step_size_overlap
                                        
        #                                     gamma = gamma * step_size
        #                                 print('given theta =\n', theta)
        #                                 print('with coords =\n', coords)
        #                                 print('dU/dtheta =\n', theta.grad)
        # #                                 print("U = ", U[None])
        #                                 print("\n")

                                        advance()
            
                                    #Plot results for current set of hyperparamaters
        
                                    print("U = ", U[None])
                                    min_coords = coords.to_numpy()
                                    plt.plot(Us)
                                    plt.ylabel('Energy')
                                    plt.xlabel('Epochs')
                                    plt.show()
                                    # min_coords = coords.to_numpy()
                                    plt.scatter(min_coords.transpose()[0], min_coords.transpose()[1])
                                    plt.plot(min_coords.transpose()[0], min_coords.transpose()[1])
                                    plt.yticks(range(-8, 8))
                                    plt.xticks(range(-8, 8))
                                    plt.xlim(-10, 10)
                                    plt.ylim(-10, 10)
                                    plt.gca().set_aspect('equal', adjustable='box')
                                    plt.title('Final State')
                                    plt.show()
                                    print(coords)
#                             break

```
